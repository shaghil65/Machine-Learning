```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv('ex1data2.txt', names = ['size', 'bedroom','price'])
```

```python
data = dataset[['bedroom','price']]
```

```python
data.head(5)
```

```python
X_df = pd.DataFrame(data['bedroom'])
y_df = pd.DataFrame(data['price'])
m = len(y_df)
```

```python
plt.figure(figsize=(8,6))
plt.plot(X_df.bedroom, y_df, 'kx')
plt.xlabel('Bedroom')
plt.ylabel('Price')
```

```python
iter = 1000
alpha = 0.01
```

```python
X_df['intercept'] = 1
```

```python
X = np.array(X_df)
y = np.array(y_df).flatten()
theta = np.array([0, 0])
```

```python
theta
```

```python
def cost_function(X, y, theta):
    m = len(y) 
    J = np.sum((X.dot(theta)-y)**2)/2/m
    return J
```

```python
cost_function(X, y, theta)
```

```python
def gradient_descent(X, y, theta, alpha, iterations):
    
    cost_history = [0] * iterations
    
    for iteration in range(iterations):
        print(X)
        print(np.shape(X)) 
        hypothesis = X.dot(theta)

        loss = hypothesis-y
        gradient = X.T.dot(loss)/m
        theta = theta - alpha*gradient
        cost = cost_function(X, y, theta)
        cost_history[iteration] = cost

    return theta, cost_history
```

```python
gd = gradient_descent(X,y,theta,alpha, iter)
```

```python
t = [gd[0][0], gd[0][1]]
best_fit_x = np.linspace(0, 5, 10)
best_fit_y = [t[1] + t[0] * xx for xx in best_fit_x]

plt.figure(figsize=(6, 6))
plt.plot(X_df, y_df, '.')
plt.plot(best_fit_x, best_fit_y, '-')
plt.xlabel('Bedroom')
plt.ylabel('Price')
plt.title('Price vs. Bedroom with Linear Regression Line')
plt.show()
```
